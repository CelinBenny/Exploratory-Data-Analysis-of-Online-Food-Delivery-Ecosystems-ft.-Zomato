Exploratory Data Analysis of Online Food Delivery Ecosystem ft.Zomato

This project analyzes factors influencing restaurant ratings,pricing and customer preferences in an onine food delivery platform using Python.

Tools Used :
 - Python
 - Pandas
 - Google Colab

IMPORT AND LOAD

```python
import pandas as pd
import numpy as np

df = pd.read_csv("zomato.csv")


```

OVERVIEW

```python
df.shape
df.head()
```

```python
df.info()
df.describe()
```

DATA QUALITY CHECK

```python
df.isnull().sum()
df.duplicated().sum()
```

CLEANING

```python
df = df.drop_duplicates()
df.columns = df.columns.str.strip()
```

```python
df['rate'] = df['rate'].str.replace('/5','')
```

```python
df['rate'] = pd.to_numeric(df['rate'],errors='coerce')
```

```python
df['approx_cost(for two people)'] = pd.to_numeric(
    df['approx_cost(for two people)'].str.replace(',',''),
    errors='coerce'
)
```

HANDLING MISSING COLUMNS

```python
df = df.dropna(subset=['rate','approx_cost(for two people)','location'])
df = df.drop(columns=['phone'])
```

DERIVED COLUMNS

```python
df['price_category'] = pd.cut(
    df['approx_cost(for two people)'],
    bins=[0,500,1000,2000,6000],
    labels=['Low','Medium','High','Luxury']
)
```

FILTERING

```python
high_rated = df[df['rate'] > 4]
high_rated.head()
```

AGGREGATION

```python
df.groupby('location')['rate'].mean()
```
